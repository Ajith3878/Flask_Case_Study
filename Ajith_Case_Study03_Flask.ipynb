Case Study on Python Flask: Using the given dataset, create a web application to take feature inputs from the user in a web page and print whether a person will purchase the product or not.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Social_Network_Ads (3).csv')
df
```

```python
#Encoding of Object Data
import sklearn.preprocessing as sk

obj_data = df.select_dtypes(include = 'object')
le = sk.LabelEncoder()
for i in obj_data:
    df[i] = le.fit_transform(df[i])
```

```python
df
```

```python
df = df.drop('User ID',axis=1)
```

```python
df
```

```python
x=df.drop(['Purchased'],axis=1)
y=df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
from sklearn import linear_model
lr = linear_model.LinearRegression()
```

```python
model=lr.fit(x_train,y_train)
```

```python
predict=model.predict(x_test)
```

```python
predict
```

```python
import pickle
pickle.dump(model,open('model.pkl','wb'))
```

```python
pickled_model = pickle.load(open('model.pkl','rb'))
```
